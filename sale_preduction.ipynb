```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
df.rename(columns={"Unnamed: 0": "Number_Index"}, inplace=True)
df.head()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#seeing the relationship Newspaper with the Sales
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Sales'], y=df['Newspaper'], data=df)
plt.xlabel('Sales')
plt.ylabel('Newspaper')
plt.title('Comparison of Column1 and Column2')
plt.show()
```

```python
#seeing the relationship Radio with the Sales
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Sales'], y=df['Radio'], data=df)
plt.xlabel('Sales')
plt.ylabel('Newspaper')
plt.title('Comparison of Column1 and Column2')
plt.show()
```

```python
#seeing the relationship TV with the Sales
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Sales'], y=df['TV'], data=df)
plt.xlabel('Sales')
plt.ylabel('Newspaper')
plt.title('Comparison of Column1 and Column2')
plt.show()
```

```python
# Split the data into train and test sets data
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#train the model 
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#make the prediction and estimate the mean error
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R-squared: {r2:.2f}")
```
